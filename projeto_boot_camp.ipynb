```python
global usuarios
usuarios = []
global contador
contador = 0
global his_acoes



def CriarExtratoDeposito(usuario_log,valor_deposito,valor_conta):
    global cont_acao
    nova_acao = {
        'num_acao' : usuario_logado['cont_acoes'],
        'usuario' :usuario_log,
        'valor_deposito':valor_deposito,
        'valor_conta': valor_conta
    }
    usuario_logado['cont_acoes'] += 1
    return usuario_logado['his_acoes'].append(nova_acao)

def CriarExtratoSacar(usuario_log,valor_saque,valor_conta):
    global cont_acao
    nova_acao = {
        'num_acao' : usuario_logado['cont_acoes'],
        'usuario' :usuario_log,
        'valor_saque':valor_saque,
        'valor_conta': valor_conta
    }
    usuario_logado['cont_acoes'] += 1
    return usuario_logado['his_acoes'].append(nova_acao)

def CriarExtratoTransferencia(usuario_log,valor_transf,usuario_recebidor):
    global cont_acao
    nova_acao = {
        'num_acao' : usuario_logado['cont_acoes'],
        'usuario' :usuario_log,
        'valor_transferencia':valor_transf,
        'usuario_recebidor' :usuario_recebidor
    }
    usuario_logado['cont_acoes'] += 1
    return usuario_logado['his_acoes'].append(nova_acao)

def VerificarCpf(cpf):
    if(len(cpf) != 11 ):
        print("--------------------------------------------")
        print("|                CPF inválido              |")
        print("--------------------------------------------")
        Aplicacao()
    else:
        return cpf

def VerificarSenha(senha):
    if(len(senha) != 4):
        print("--------------------------------------------")
        print("|               SENHA inválida             |")
        print("--------------------------------------------")
        Aplicacao()
    else:
        return senha

def SelecaoAcoes():
  try:
      print("--------------------------------------------")
      print(f"|  Seja Bem Vindo {usuario_logado['nome']}|")
      print("--------------------------------------------")
      print("|         Escolha uma Ação                 |")
      print("--------------------------------------------")
      print("| 1 - Depositar                            |")
      print("| 2 - Sacar                                |")
      print("| 3 - Transferir                           |")
      print("| 4 - Extrato                              |")
      print("| 5 - Menu inicial                         |")
      print("--------------------------------------------")
      acao = int(input("Digite a acão desejada: "))
      if(acao == 1):
          Depositar()
      elif(acao == 2):
          Sacar()
      elif(acao == 3):
          Transferir()
      elif(acao == 4):
          Extrato()
      elif(acao == 5):
          MenuInicial()
      else:
        print("opção inválida tente novamente")
        Aplicacao()
  except TypeError as t_err:
    print("tipo de entrada inválida tente novamente")
    Aplicacao()



def Registrar():
    global contador
    print("--------------------------------------------")
    print("|                CADASTRO                  |")
    print("--------------------------------------------")
    print("|DADOS: NOME*, CPF*, SENHA*                |")
    print("|OBS: Senha deve ser um pin de 4 dígitos   |")
    print("--------------------------------------------")
    nome = input("|Nome: ")
    cpf = input("|CPF: ")
    senha = input("|SENHA: ")
    VerificarCpf((cpf))
    VerificarSenha((senha))
    usuario = {
            "id_usuario": contador,
            "nome":  str(),
            "cpf":   str(),
            "senha": str(),
            "conta": float(),
            "cont_acoes" : 0,
            "his_acoes": []
        }

    novo_usuario = usuario.copy()
    novo_usuario['nome'] = str(nome.upper())
    novo_usuario['cpf'] = str(cpf)
    novo_usuario['senha'] = str(senha)
    usuarios.append(novo_usuario)
    contador += 1
    if(len(usuarios) >= 1):
        Logar()
    else:
        print("--------------------------------------------")
        print("|            ERRO REINICIANDO...           |")
        print("--------------------------------------------")
        Aplicacao()

def Logar():
    print("--------------------------------------------")
    print("|               LOGIN                     |")
    print("--------------------------------------------")
    print("| ENTRE COM O SEU: CPF E SENHA            |")
    print("--------------------------------------------")
    global usuario_logado
    cpf = input("| Digite o seu CPF: ")
    senha = input("| Digite a sua senha: ")

    for usuario in usuarios:
        if(usuario['cpf'] == cpf and usuario['senha'] == senha):
            usuario_logado = usuario
            SelecaoAcoes()
            return

    print("--------------------------------------------")
    print("|          Dados Incorretos                |")
    print("--------------------------------------------")
    opcao_l = str(input("| Escolha 1 para tentar logar\n| Escolha 2 para se Registrar: "))
    if(opcao_l == '1'):
        Logar()
    elif(opcao_l == '2'):
        Registrar()


def Depositar():
    print("--------------------------------------------")
    print("|             DEPOSITAR                    |")
    print("--------------------------------------------")
    print("| DIGITE O VALOR A SER DEPOSITADO:         |")
    print("--------------------------------------------")
    valor_deposito = float(input())
    usuario_logado['conta'] += valor_deposito
    print("--------------------------------------------")
    print(f"| Depósito realizado. Novo saldo: {usuario_logado['conta']}   |")
    print("--------------------------------------------")

    usuario_log = usuario_logado['nome']
    valor_conta = usuario_logado['conta']

    CriarExtratoDeposito(usuario_log, valor_deposito, valor_conta)

    voltar_menu = ""
    voltar_menu = str(input("Digite algo para voltar para o menu"))
    if(len(voltar_menu) > 0):
        SelecaoAcoes()


def Sacar():
    print("--------------------------------------------")
    print("|                SACAR                     |")
    print("--------------------------------------------")
    print("| DIGITE O VALOR A SER SACADO:             |")
    print("--------------------------------------------")
    valor_saque = float(input())
    if(usuario_logado['conta'] >= valor_saque):
        usuario_logado['conta'] -= valor_saque
        print("--------------------------------------------")
        print(f"| Saque realizado. Novo saldo: {usuario_logado['conta']}   |")
        print("--------------------------------------------")
    else:
        print("--------------------------------------------")
        print("|          Saldo insuficiente              |")
        print("--------------------------------------------")

    usuario_log = usuario_logado['nome']
    valor_conta = usuario_logado['conta']

    CriarExtratoSacar(usuario_log, valor_saque, valor_conta)

    voltar_menu = ""
    voltar_menu = str(input("Digite algo para voltar para o menu"))
    if(len(voltar_menu) > 0):
        SelecaoAcoes()


def Transferir():
    print("--------------------------------------------")
    print("|             TRANSFERIR                   |")
    print("--------------------------------------------")
    print("| DIGITE PARA QUEM VOCÊ QUER TRANSFERIR    |")
    print("| USE O CPF                                |")
    print("--------------------------------------------")
    usuario_recebidor = str(input("| CPF do destinatário: "))
    print("--------------------------------------------")
    print("| DIGITE A QUANTIA QUE SERÁ TRANSFERIDA    |")
    print("--------------------------------------------")
    valor_transf = float(input())

    if(usuario_recebidor == usuario_logado['cpf']):
        print("--------------------------------------------")
        print("| Você não pode transferir para sua própria|")
        print("| conta.                                   |")
        print("--------------------------------------------")
        SelecaoAcoes()
    else:
        for usuario in usuarios:
            if usuario_recebidor == usuario['cpf']:
                if usuario_logado['conta'] >= valor_transf:
                    usuario_logado['conta'] -= valor_transf
                    usuario['conta'] += valor_transf
                    print("--------------------------------------------")
                    print(f"| Transferência de {valor_transf} realizada para {usuario['nome']}   |")
                    print("--------------------------------------------")
                    usuario_log = usuario_logado['nome']
                    CriarExtratoTransferencia(usuario_log,valor_transf,usuario_recebidor)
                else:
                    print("--------------------------------------------")
                    print("|          Saldo insuficiente              |")
                    print("--------------------------------------------")
                voltar_menu = ""
                voltar_menu = str(input("Digite algo para voltar para o menu"))
                if(len(voltar_menu) > 0):
                  SelecaoAcoes()


def Extrato():
    try:
        print("--------------------------------------------")
        print("|               EXTRATO                    |")
        print("--------------------------------------------")
        print(f"| VALOR EM CONTA: {usuario_logado['conta']}                |")
        print("--------------------------------------------")
        print("|VOCÊ DESEJA RETIRAR O SEU EXTRATO \n|DIGITE 1 PARA SIM OU 2 PARA VOLTAR")
        opcao_e = str(input(""))
        if(opcao_e == '1'):
            for acao in usuario_logado['his_acoes']:
                print(acao)
            voltar_menu = ""
            voltar_menu = str(input("Digite algo para voltar para o menu"))
            if(len(voltar_menu) > 0):
              SelecaoAcoes()
        elif(opcao_e == '2'):
            print("voltando ao menu...")
            SelecaoAcoes()
    except:
        print("--------------------------------------------")
        print("|            Erro, tente novamente         |")
        print("--------------------------------------------")
        Extrato()

def MenuInicial():
    print("Saindo...")
    Aplicacao()


def Aplicacao():
    print("--------------------------------------------")
    print("|   Seja Bem Vindo ao Banco Cesar          |")
    print("--------------------------------------------")
    print("|         Escolha uma opção                |")
    print("--------------------------------------------")
    print("| 1 - Criar Conta                          |")
    print("| 2 - Logar                                |")
    print("--------------------------------------------")
    opcao = str((input("Digite a opção desejada: ")))

    if(opcao == '1'):
        Registrar()
    elif(opcao == '2'):
        Logar()

Aplicacao()
```
